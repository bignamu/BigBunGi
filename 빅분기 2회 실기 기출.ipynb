# 작업형 1-2
### 주어진 데이터 첫번째 행 부터 순서대로 80%까지의 데이터를 추출 후 'total_bedrooms' 변수의 결측값(NA)을 'total_bedrooms' 변수의 중앙값으로 대체하고 대체 전의 'total_bedrooms' 변수 표준편차값과 대체 후의 'total_bedrooms' 변수 표준편차 값 산출

```python
df12 = pd.read_csv('housing.csv')
```

```python
df12 = df12[:int(len(df12)*0.8)]
```

```python
df12.isnull().sum()
```

```python
bedroom_medi = df12['total_bedrooms'].median()
bedroom_medi
```

```python
after = df12.fillna(bedroom_medi)
df12.head()
```

```python
after.isnull().sum()
```

```python
abs(after['total_bedrooms'].std() - df12['total_bedrooms'].std())
```
